This file categorize the `location_category` field of an xlsx into world regions.

```python

import sys, os

sys.path.append(os.path.abspath(os.path.join("..", "..", "src")))

import time
import json

import pandas as pd
import numpy as np
from tqdm import tqdm
tqdm.pandas()

from common.app import App

app_run = App(debug=True)
```

```python
filename = "agents_full_year1_COVID19.xlsx"
xlsx_path = os.path.join(app_run.root_dir, "src", "resources", filename)

df = pd.read_excel(xlsx_path, header=0, keep_default_na=False)
```

```python
df
```

```python
regions_filename = "countries_regions.json"
regions_path = os.path.join(app_run.root_dir, "src", "resources", regions_filename)
with open(regions_path, "r") as fjson:
    regions = json.load(fjson) 
```

```python
def categorize_location(location_cat):
    location_cat = location_cat.strip()
    if location_cat == "Europe":
        return "Europe"
    elif location_cat == "International":
        return "International"
    elif location_cat == "NA":
        return "NA"
    elif location_cat == "TweepError":
        return "TweepError"
    else:
        if location_cat in regions.keys():
            return regions[location_cat] 
        else:
            return None
```

```python
df["region"] = df["location_category"].progress_apply(categorize_location)
```

```python
# Output df

output_f = "agents_full.xlsx"
out_path = os.path.join(app_run.root_dir, "src", "resources", output_f)
df.to_excel(out_path, index=False, engine="openpyxl")
```
